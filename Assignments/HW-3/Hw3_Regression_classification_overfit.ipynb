**Question 2.1** Suppose you are building a model to predict whether a person has a rare disease that only affects 1% of the population. You have collected a dataset of 10,000 people, of which only 100 have the disease. You train your model and get an accuracy of 99%, which seems very high. However, you are concerned that the model may not be performing well on the minority class. What evaluation metric would you use to assess the model's performance, and why?

I would use Mean Square Error. Since since is an easy way to compute the gradient, where the difference of actual and computed squared divided with total number of data shows the difference in error while magnifying the outlying instances, hence in this case the minority class we expect the model might not be performing well. As a result, one can quickly identify which classes are not performing with high accuracy.


**Question 2.2** What evaluation metric would be most appropriate for a machine learning model that is used to screen patients for a certain disease in order to refer them for further testing, and why? Would you use precision or recall as the primary evaluation metric for this problem, and why?


I would use Area Under the Curve. Area Under the Curve includes sensitivity, specificity, and false positive rates Since this scenario is a test of categorizing someone whether or not they have a given disease x, there is a chance that the model can correctly predict yes (true positive rate or sensitivity) or no (true negative rate), and incorrectly predict yes (false positive rate). Thus, since all of this possible outcomes can be evaluated using this method, I would choose confusion matrix to determine if model screens patients accurately.

I would not use precision or recall to evaluate the model in this example, since these parameters only check how many instances were correctly classified. This would mean that since incorrect outputs aren't given an emphasis, people who should have been directed to further testing might miss their opportunity. Thus, a metrics that checks both the positives and negatives is needed.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('car_dataset.csv')
df.head()


```

```python
# Check for duplicate values and drop them
print("Checking for duplicate values: ", df.duplicated().sum() != 0)

# Pre processing data
df.describe()
```

```python
# Remove text only information columnx
df_c = df.copy()
df_1 = df_c.drop(['Unnamed: 0', 'make', 'model', 'seller_type'], axis=1)

# Convert the Type column Low,Medium,High to 1,2,3
df_1['price_category'] = df_1['price_category'].apply(lambda x: 1 if x == 'Low' else 2 if x == 'Medium' else 3)

# Convert Owner info to first, second, third, more to 1, 2, 3, 4
df_1['owner'] = df_1['owner'].apply(lambda x: 1 if x == 'First Owner' else 2 if x == 'Second Owner' else 3 if x == 'Third Owner' else 4)

# For transmission Manual , Automatic to 1, 2
df_1['transmission'] = df_1['transmission'].apply(lambda x: 1 if x == 'Manual' else 2)

# For fuel Petrol, Diesel, LPG to 1, 2, 3
df_1['fuel'] = df_1['fuel'].apply(lambda x: 1 if x == 'Petrol' else 2 if x == 'Diesel' else 3)

df_1.head()
```

```python
# Check for missing values
print(df_1.isnull().sum())
```

```python
# Print count of Unique values of each column
for col in df_1.columns:
    print(col, df_1[col].nunique())

```

```python
# Remove units from parameters

variable = ['engine','max_power']

for cmn in variable:
   df_1[cmn] = df_1[cmn].str.strip()
   if cmn == 'engine':
    df_1[cmn] = df_1[cmn].str.replace("CC","")
   else:
     df_1[cmn] = df_1[cmn].str.replace("bhp","")
df_1.head()
```

```python
df_1['year'] = df_1['year'].astype(float)
df_1['transmission'] = df_1['transmission'].astype(float)
df_1['fuel'] = df_1['fuel'].astype(float)
df_1['engine'] = df_1['engine'].astype(float)
#df_1['max_power'] = df_1['max_power'].astype(float)
df_1['max_power'] = pd.to_numeric(df_1["max_power"], downcast="float")
df_1['seats'] = df_1['seats'].astype(float)
df_1['km_driven'] = df_1['km_driven'].astype(float)
df_1['price_category'] = df_1['price_category'].astype(float)

# Replace empty cells with column mean

variable = ['engine','max_power','seats']

means = df_1.mean()

print(means)
```

```python
# Replace empty cells with column mean

variable = ['engine','max_power','seats']
a = 4
for j in range(len(variable)):
  a += j
  for i in range(len(df_1[variable[j]])):
    if pd.isnull(df_1.loc[i, variable[j]]):
      df_1.loc[i, variable[j]] = means[a]

# Check for missing values
print(df_1.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split

df_c = df_1.copy()
x = df_c.drop('price_category',axis=1)
y = df_1['price_category']

# Split the data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=21)

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Train a Logistic Regression model

# lr : Logistic Regression model  , y_pred_lr : Predicted values of y_test

lr = LogisticRegression(solver='liblinear', random_state=0)
lr.fit(X_train,y_train)
y_pred_lr = lr.predict(X_test)
y_pred_prob = lr.predict_proba(X_test)
# Evaluation metrics

y_pred_lr = lr.predict(X_test)
log_train = round(lr.score(X_train, y_train) * 100, 2)
log_accuracy = round(accuracy_score(y_pred_lr, y_test) * 100, 2)
```
